```python
class node:
  def __init__(self,data):
    self.addr=None
    self.data=data


class stack:
  def __init__(self):
    self.head=None
  
  def empty(self):
    if self.head==None:
      return True
    else:
      return False

  def push(self,data):
    if self.empty():
      self.head=node(data)
    else:
      newnode = node(data)       #head acts as address flag
      newnode.addr=self.head
      self.head=newnode

  def pop(self):
    if self.empty():
      print("Underflow.")
    else:
      self.head=self.head.addr
      print("Top popped from element")
  def traverse(self):
    flag=self.head
    if self.head==None:
      print("Empty stack")
    else:
      while(flag!=None):
        print(flag.data)
        flag=flag.addr

  def check(self):
    count=0
    if self.head!=None:
      for i in self:
        count+=1
      return count
  
  def top(self):
    if self.head==None:
      print("empty stack")
    else:
      print(self.head.data)
  
  def destroy(self):
    del self

def usechoice(choice,obj):
  while(choice>0 and choice<6):
    if choice==1:
      k=input("Enter data for stack:")
      obj.push(k)
    elif choice==2:
      obj.pop()

    elif choice==3:
      obj.top()
    elif choice==4:
      obj.traverse()
    elif choice==5:
      obj.destroy()
      print("Stack destroyed program terminated")
      break
    else:
      break
    choice=int(input('Enter your choice for stack:\n(1.PUSH\t2.POP\t3.TOP\t4.TRAVERSE\t5.DESTROY\t6.EXIT):'))

stack1=stack()
i=int(input('Enter your choice for stack:\n(1.PUSH\t2.POP\t3.TOP\t4.TRAVERSE\t5.DESTROY\t6.EXIT):'))
usechoice(i,stack1)
```
